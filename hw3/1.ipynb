```python
import pandas as pd
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def get_data(file_name):
    data = pd.read_csv(file_name).values
    return data[:, 0: -1].astype('float64'), data[:, -1].astype('int')
```

```python
def svm_classifier(points, labels, ker, deg):
    matplotlib.rcParams['figure.figsize'] = [10, 10]
    clf = svm.SVC(kernel=ker, degree=deg)
    xs = points[:, 0]
    ys = points[:, 1]
    clf.fit(points, labels)
    xx, yy = np.meshgrid(np.arange(xs.min() - 1, xs.max() + 1, 0.02),
                         np.arange(ys.min() - 1, ys.max() + 1, 0.02))
    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    plt.contourf(xx, yy, z, cmap=plt.cm.Set2, alpha=0.5)
    plt.scatter(xs, ys, c=labels, cmap=plt.cm.Set2, s=20, edgecolors=None)
    support_x = xs[clf.support_]
    support_y = ys[clf.support_]
    support_labels = labels[clf.support_]
    plt.scatter(support_x, support_y, c=support_labels, cmap=plt.cm.Set2, s=20, edgecolors='black')
    plt.show()
```

```python
points, labels = get_data("datasets/blobs2.csv")
```

```python
svm_classifier(points, labels, 'linear', 0)
```

```python
svm_classifier(points, labels, 'poly', 2)
```

```python
svm_classifier(points, labels, 'poly', 3)
```

```python
svm_classifier(points, labels, 'poly', 5)
```

```python
svm_classifier(points, labels, 'rbf', 0)
```
